#### Reformatting the Estimated Effect Duration column values into respected bins. Then adding two additional columns such that the upper and lower time is calculated from its values. 

```python
df.Version_Duration.unique()
```

```python
df.Estimated_Effect_Duration.value_counts()
```

```python
# cleaning up values where it's a range of numbers instead of with strings
df['Estimated_Effect_Duration'] = df['Estimated_Effect_Duration'].map({'4 to 8 hours': '4-8',
    'More than 8 hours': '8+',
    '2 to 4 hours': '2-4',
    'Less than 2 hours': '<2'})
```

```python
#converting hour range into numeric numbers with upper and lower bounds of the bins 
upper  = {'4-8': 8,
         '2-4': 4,
         '8+': 12,
         '<2': 2}
df['Upper_Effect_Duration'] = df['Estimated_Effect_Duration'].map(upper)


lower = {'4-8': 4,
        '2-4':2, 
        '8+': 8,
        '<2': 0}
df['Lower_Effect_Duration'] = df['Estimated_Effect_Duration'].map(lower)
```
