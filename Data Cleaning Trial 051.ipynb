```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('Trial_051_deident.csv', parse_dates=['Start Date', 'End Date'])
df.head()
```

```python
df.info()
```

```python
df = df.rename(columns={'Start Date': 'Start_Date',
                  'End Date': 'End_Date',
                  'First Name': 'Last_Name',
                  'For how long have you tried this version?' : 'Version_Duration',
                  'Sleep_Quality' : 'Sleep_Quality',
                  'Overall, your experience with this product so far is': 'Overall_Experience',
                  'Focus/ Attention/ Concentration - ': 'Focus_Atten_Concen',
                  'Memory (short and long term) - ': 'Memory',
                  'Drive/ Passion/ Motivation - ': 'Drive_Passion_Motivation',
                  'Physical Energy - ': 'Physical_Energy',
                  'Capacity/ productivity - ': 'Capacity_Productivity',
                  'Decisiveness - ': 'Decisiveness',
                  'Verbal fluency / word recall - ': 'Verbal_Fluency_Word_Recall',
                  'Creativity / insight - ': 'Creativity_Insight',
                  'Interpersonal capability / empathy - ': 'InterpersonalCapability_Empathy',
                  'Emotional stability - ': 'Emotional_Stability',
                  'Euphoria / joy / happiness - ': 'Euhphoria_Joy_Happiness',
                  'Estimated_Effect_Aprox' : 'Estimated_Effect_Duration',
                  'If you are a current user of the Drug, how would you compare this formula?': 'Current_User_Comparison'})
```

```python
df.head()
```

```python
df.describe()
```

```python
df.Estimated_Effect_Duration.value_counts()
```

```python
df['Estimated_Effect_Duration'] = df['Estimated_Effect_Duration'].map({'4 to 8 hours': '4-8',
    'More than 8 hours': '8+',
    '2 to 4 hours': '2-4',
    'Less than 2 hours': '<2'})
```

```python
#converting hour range into numeric numbers with upper and lower bounds of the bins 
upper  = {'4-8': 8,
         '2-4': 4,
         '8+': 12,
         '<2': 2}
df['Upper_Effect_Duration'] = df['Estimated_Effect_Duration'].map(upper)


lower = {'4-8': 4,
        '2-4':2, 
        '8+': 8,
        '<2': 0}
df['Lower_Effect_Duration'] = df['Estimated_Effect_Duration'].map(lower)
```

```python
df.head()
```

```python
df.Version_Duration.value_counts()
```

```python
week = '1 week ' + '(' + '5 days' + ')'

df['Version_Duration'] = df['Version_Duration'].map({ week : '1 weeks', '1 day': '1 day', '2 weeks':'2 weeks', '3 weeks': '3 weeks'})
```

```python
df.head()
```

```python
df.Current_User_Comparison.value_counts()
```

```python
user = {'I am not a XNootropic user': 'Non-user',
        'This is less good than XNootropic':'XNootropic User',
        'This is better than XNootropic': 'XNootropic User',
        'This is about the same as XNootropic' : 'XNootropic User'}


df['User_Status'] = df['Current_User_Comparison'].map(user)
```

```python
df.head()
```

```python
#rearranging the column order so that they match
cols = df.columns.tolist()
col = cols[0:18] + cols[19:21] + cols[18:19] + cols[21:]
df = df[col]
```
