```python
import pandas as pd
```

```python
df = pd.read_csv('Trial_099_deident.csv')
```

```python
df.head()
```

```python
renaming = {'I am not a Qualia user': 'I am not a XNootropic user',
            'This is less good than Qualia':'This is less good than XNootropic',
            'This is better than Qualia': 'This is better than XNootropic', 
            'This is about the same as Qualia' : 'This is about the same as XNootropic'}
```

```python
df['If you are a current user of the Drug, how would you compare this formula?'] = df['If you are a current user of the Drug, how would you compare this formula?'].map(renaming)
```

```python
df.head()
```

```python
df.to_csv('Trial_099_deident.csv', index=False)
```
